```python
import numpy as np
from sklearn.model_selection import train_test_split
from data_pre import *
from PMF import *
```

```python
file = 'Data/datasets/ml-100k/u.data' 
user_c,item_c, rated_info, rated_items = data_propress(file)

train_data, test_data = train_test_split(rated_info, test_size=0.3)
print('train num: {}        test num: {}'.format(len(train_data), len(test_data)))
print()
print("start training...")
print()
model = PMF(train_data, 
            test_data, 
            user_c, 
            item_c,
            K=10,
            learning_rate=0.005, 
            lamda1=0.1,
            lamda2=0.1, 
            max_iter=100)
U, V, results = model.train()
```
